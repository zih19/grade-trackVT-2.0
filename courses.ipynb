Aerospace and Ocean Engineering(AOE)

```python
AOE_CSV_files = os.listdir(directory_path_AOE)
for file in AOE_CSV_files:
    AOE_file_path = os.path.join(directory_path_AOE, file)
    df = pd.read_csv(AOE_file_path)
    print("Basic Profile: ")
    print("\n")
    print("The course number of the AOE course is: ", df['Course No.'].unique())
    print("\n")
    print("The course title is: ", df['Course Title'].unique())
    print("\n")
    print("The average gpa for this course is: ", round(df['GPA'].mean(), 2))
    print("\n")
    print("The total number of people to be enrolled is: ",  df['Graded Enrollment'].sum())
    print("\n")
    print("The total number of people to be withdrawn is: ", df['Withdraws'].sum())
    print("\n")
    print("The credit(s) for this course is: ", df['Credits'].unique())
    print("\n")
    print("The name(s) of the instructors are:  ", df['Instructor'].unique())
    print("\n")

    print("\n")
    print("People Profile: ")
    print("\n")
    print("The total number of students receiving A is: ", round((df['Graded Enrollment'] * (df['A (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving A- is: ", round((df['Graded Enrollment'] * (df['A- (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving B+ is: ", round((df['Graded Enrollment'] * (df['B+ (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving B is: ", round((df['Graded Enrollment'] * (df['B (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving B- is: ", round((df['Graded Enrollment'] * (df['B- (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving C+ is: ", round((df['Graded Enrollment'] * (df['C+ (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving C is: ", round((df['Graded Enrollment'] * (df['C (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving C- is: ", round((df['Graded Enrollment'] * (df['C- (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving D+ is: ", round((df['Graded Enrollment'] * (df['D+ (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving D is: ", round((df['Graded Enrollment'] * (df['D (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving D- is: ", round((df['Graded Enrollment'] * (df['D- (%)'] / 100)).sum(), 0))
    print("\n")
    print("The total number of students receiving F is: ", round((df['Graded Enrollment'] * (df['F (%)'] / 100)).sum(), 0))
    print("\n")

    print("\n")
    print("Proportion Profile: ")
    print("\n")
    proportion_A = (df['Graded Enrollment'] * df['A (%)'] + df['Graded Enrollment'] * df['A- (%)']).sum() / df['Graded Enrollment'].sum()
    print("The total percent of students at A is: ", round(proportion_A, 2))
    print("\n")
    proportion_B = (df['Graded Enrollment'] * df['B+ (%)'] + df['Graded Enrollment'] * df['B (%)'] + df['Graded Enrollment'] * df['B- (%)']).sum() / df['Graded Enrollment'].sum()
    print("The total percent of students at B is: ", round(proportion_B, 2))
    print("\n")
    proportion_C = (df['Graded Enrollment'] * df['C+ (%)'] + df['Graded Enrollment'] * df['C (%)'] + df['Graded Enrollment'] * df['C- (%)']).sum() / df['Graded Enrollment'].sum()
    print("The total percent of students at C is: ", round(proportion_C, 2))
    print("\n")
    proportion_D = (df['Graded Enrollment'] * df['D+ (%)'] + df['Graded Enrollment'] * df['D (%)'] + df['Graded Enrollment'] * df['D- (%)']).sum() / df['Graded Enrollment'].sum()
    print("The total percent of students at D is: ", round(proportion_D, 2))
    print("\n")
    proportion_F = (df['Graded Enrollment'] * df['F (%)']).sum() / df['Graded Enrollment'].sum()
    print("The total percent of students at F is: ", round(proportion_F, 2))
    print("\n")
```
